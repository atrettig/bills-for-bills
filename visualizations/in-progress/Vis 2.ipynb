```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trades_df = pd.read_csv('politician-trading-data-BUY2020.csv', usecols=['REP_FULLNAME', '% Change 1 Year'])

print(sum(trades_df['REP_FULLNAME'].isna()))
print(sum(trades_df['% Change 1 Year'].isna()))

trades_df = trades_df[trades_df['REP_FULLNAME'].notna()]
trades_df = trades_df[trades_df['% Change 1 Year'].notna()]

print(sum(trades_df['REP_FULLNAME'].isna()))
print(sum(trades_df['% Change 1 Year'].isna()))
```

```python
#dic = {'David Roe': [29, ##, ##, etc.]}

dic = {}
update_list = []

for name, change in zip(trades_df['REP_FULLNAME'], trades_df['% Change 1 Year']):
    #print(name)
    if name not in dic:
        dic[name] = [int(change.strip("%"))]
    elif change != '#DIV/0!':
        update_list = dic.get(name)
        update_list.append(int(change.strip("%")))
        dic.update({name: update_list})
print(dic)
```

```python
top = {}
#top['Thomas Suozzi'] = 0
for name, change in zip(trades_df['REP_FULLNAME'], trades_df['% Change 1 Year']):
    #print(name)
    if name not in top:
        top[name] = int(str(change).strip("%"))
    elif change != '#DIV/0!':
        average = top.get(name)
        change = int(str(change).strip("%"))
        #print(dic.get(name))
        average += change
        top.update({name: average})
print(top)
```

```python
num = {}

for name in trades_df['REP_FULLNAME']:
    if name not in num:
        num[name] = 1
        
    else:
        get_name = int(num.get(name))
        get_name += 1
        #print(get_name)
        num.update({name: get_name})
print(num)
```

```python
answer = {}

for name, numStocks, total_percentage in zip(num, num.values(), top.values()):
    #print(numStocks)
    #print(total_percentage)
    answer[name] = (total_percentage/numStocks)
#print(answer)

answer = dict(sorted(answer.items(), key=lambda item: item[1], reverse=True))
answer
```

```python
labels, data = [*zip(*dic.items())]  # 'transpose' items to parallel key, value lists

# or backwards compatable    
labels, data = dic.keys(), dic.values()

plt.boxplot(data)
plt.xticks(range(1, len(labels) + 1), labels)
plt.show()
```
